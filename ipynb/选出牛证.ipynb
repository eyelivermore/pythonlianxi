```python
import futuquant as ft
import pandas as pd
import numpy as np
```

```python
def codeshu():#找到牛证
    ok,kz= quote_ctx.get_stock_basicinfo(ft.Market.HK, ft.SecurityType.WARRANT) 
    #选出牛证BULL 熊证BEAR
    dm = kz[(kz.stock_owner =='HK.800000')&(kz.stock_child_type=='BULL')].code
    print('总共有',len(dm),'牛证')
    list(dm)
    print(type(dm))
    return list(dm)
```

```python
def dircode(quote_ctx,code,cj):
    #找到大于200,150的,i指的是差价
    ret,data = quote_ctx.get_market_snapshot(code)#市场快照
    if ret != 0:
        print('错误信息',data)
    else:
       # print(data)
        data = data[data.wrt_recovery_price - data.wrt_strike_price == cj]
        #行使价减去回收价
        data = data[data.suspension == False]
        datas = data[data.wrt_conversion_ratio==10000.0].loc[:,['code','last_price','wrt_strike_price','wrt_recovery_price','wrt_conversion_ratio']].copy()
       
        return datas
```

```python
quote_ctx = ft.OpenQuoteContext(host='127.0.0.1', port=11111)
chajia = 200    #行驶价加收价之间的价差
df = pd.DataFrame(columns = [])
ls = codeshu()
#循环每次取列表中的200个给dircode函数
for i in range(0,len(ls),200):#
    print(i)    
    aa = dircode(quote_ctx,code=ls[i:i+200],cj = chajia)
    #把符合的添加到df这个Dataframe中
    df = df.append(aa.copy())
#重新编排一下columns
df.columns = ['代码','价格','行使价','回收价','换股比例'] 
#过滤掉价格等于0的
df = df[df['价格']>0]
#重新编排一下index
df.index = range(len(df['价格'].values))
df.sort_values(by='价格')#以价格排序


quote_ctx.close()
```
