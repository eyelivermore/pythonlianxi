```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
```

```python
def page_auto(url,i):#自动翻页
    browser.get(url=rul)
    idtextss = WebDriverWait(browser,10)#等待10秒
    try:
        element = idtextss.until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#feedList > div:nth-child(1)")
        ))#判断第一条是不是送种子的
        htmls = browser.page_source
        #print("得到ID:",i)
        return htmls
        
    except:
        pass
```

```python
def fenxirul(htmls,i):
    zhongzibao = re.findall(r'<a href=\"https://seed\.futunn\.com/package\?key=.*?">(.*?)</a',htmls)
    zhongzititle = re.findall(r'png\"><span class=\"title\">(.*?)请笑纳\！</span>',htmls)
    zhongzi = re.findall(r'<a class=\"orgLink\" href=\"(.*?)\" target=\"_blank\">',htmls)
    times = re.findall(r"<a target=\"_blank\" href=\"/nnq/detail.*?\">(.*?)</a>",htmls)
    if len(times[0])==4:
        time = times[1]
    else:
        time = times[0]
    if zhongzibao !=[]:
        print(time,i,"种子包",'\n',zhongzibao[0])
    

    elif zhongzititle!=[]:
        print(time,i,zhongzititle[0],"\n",zhongzi[0])
    
```

```python
htmls = open('D:/pyexe/id.csv',"r").read()
url1 = "https://www.futunn.com/account/nnq?uid="
browser = webdriver.Firefox()
#browser = webdriver.Chrome()
for i in htmls.split(','):
    if i=="":
        print('结束')
        break
    rul = url1 + i
    html = page_auto(rul,i)
    fenxirul(html,i)

```
