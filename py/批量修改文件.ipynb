```python
import pandas as pd
import numpy as np
```

```python
st = pd.read_excel('D:\\处理文件.xlsx')
```

```python
st.insert(loc = 0,column='ID',value='')
st.insert(loc = 2,column='专业',value='二建VIP专辅定制班')
st.insert(loc = 3,column='课程',value='二建：公路工程管理与实务')
st.insert(loc = 4,column='章节',value='超级精讲↓↓↓，')
st.insert(loc = 5,column='难度',value='')
st.insert(loc = 11,column='答案选项6',value='')
st.insert(loc =13,column='正确答案',value='')

d = {'A':'题干','B':'答案选项1',
    'C':'答案选项2','D':'答案选项3',
    'E':'答案选项4','F':'答案选项5',
    'H':'试题讲解'}
st.rename(columns=d,inplace=True)

def thzf(s):
    b=''
    c = ''
    for i in s[4:]:
        if i=='A':
            c = '1，'
        if i == 'B':
            c = '2，'
        if i == 'E':
            c = '5，'
        if i == 'C':
            c = '3，'
        if i == 'D':
            c = '4，'
        b=b+c
    b=b.rstrip('，')
    return b


st['正确答案'] = st.G.apply(thzf)

del st['G']


st['题干']=st['题干'].apply(lambda x:str(x).replace('【单选题】',''))
st['答案选项1'] = st['答案选项1'].apply(lambda x:str(x).replace('A.',''))
st['答案选项2'] = st['答案选项2'].apply(lambda x:str(x).replace('B.',''))
st['答案选项3'] = st['答案选项3'].apply(lambda x:str(x).replace('C.',''))
st['答案选项4'] = st['答案选项4'].apply(lambda x:str(x).replace('D.',''))
st['试题讲解'] = st['试题讲解'].apply(lambda x:str(x).replace('【解析】',''))
st['答案选项5'] = st['答案选项5'].apply(lambda x:str(x).replace('E.',''))
st['题干'] = st['题干'].apply(lambda x:str(x).replace('【多选题】',''))
st['难度'] = st['难度'].apply(lambda x:np.random.randint(2,5)) 
st['答案选项5'] = st['答案选项5'].apply(lambda x:str(x).replace('nan',''))


df = pd.DataFrame(columns=['ID', '题干', '专业', '课程', '章节', '难度', '答案选项1', '答案选项2', '答案选项3', '答案选项4','答案选项5', '答案选项6', '正确答案', '试题讲解'])
```
